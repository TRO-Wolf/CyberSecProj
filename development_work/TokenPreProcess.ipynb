```python
class SimpleTokenizer:
    def __init__(self):
        self.token_to_id = defaultdict(lambda: len(self.token_to_id))
        self.token_to_id['<PAD>'] = 0  # Padding token

    def tokenize(self, text):
        # Simple tokenization by splitting on non-word characters
        tokens = re.findall(r'\w+|\S', text)
        return tokens

    def convert_tokens_to_ids(self, tokens):
        return [self.token_to_id[token] for token in tokens]

    def tokenize_column(self, series):
        # Apply tokenization to each row in the pandas series
        tokenized = series.apply(self.tokenize)
        return tokenized
```

```python
class ZipDataMerger:
    def __init__(self):
        self.compressed_files_list = [
            'compressed_chunk_1.zip', 'compressed_chunk_2.zip', 'compressed_chunk_3.zip'
        ]
        self.base_files_list = [
            'chunk_1.csv', 'chunk_2.csv', 'chunk_3.csv'
        ]

    
    def read_data(self):
        self.df_list = []

        for compressed_file, base_file in zip(self.compressed_files_list, self.base_files_list):
            with zipfile.ZipFile(compressed_file, 'r') as zipf:
                with zipf.open(base_file) as csv_file:
                    df = pd.read_csv(csv_file)
                    self.df_list.append(df)

```
